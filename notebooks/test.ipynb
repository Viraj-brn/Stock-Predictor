```python
import torch
from datetime import date
from dateutil.relativedelta import relativedelta
import yfinance as yf
```

```python
end_date = date.today()
start_date = end_date - relativedelta(years=6)
```

```python
df = yf.download(["MSFT", "JNJ", "PEP", "JPM", "XOM"], start=start_date, end=end_date)
```

```python
df.shape
df.columns
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[['Close']])
```

```python
import joblib
joblib.dump(scaler, "../data/close_scaler.pkl")
```
